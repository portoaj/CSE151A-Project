```python
!rm -rf runs
from ultralytics import YOLO

# Code sourced/ modified from the github page: https://github.com/ultralytics/ultralytics
# And from the docs: https://docs.ultralytics.com/modes/train/#train-settings


model = YOLO("yolov8s.pt")
model.train(data="/content/AnnotatedDataYoloFormat/data.yaml", epochs=100, batch=8, plots=True, seed=0, patience=25)
```

```python
from PIL import Image

# Printing the training/ val loss and accuracy graphs
display(Image.open('runs/detect/train/results.png'))
```

## Evaluating the Train/ Loss Graph

These graphs were very impressive to me especially in comparison to the train/ loss graphs on the previous two DETR models. These models achieved close to perfect (and sometimes) perfect recall as well as an AP50 of allmost 1 as well so almost every label was correctly predicted and the placement was very close.

The validation loss seems to follow the training loss for each loss type (class, box and focal) so I believe that the model is not overfitting since there is no divergence towards the end of the model training. Additionally, I don't think there is underfitting since the model performs very well on both the training and validation sets as discussed above. Thus, the model seems to be balanced.
